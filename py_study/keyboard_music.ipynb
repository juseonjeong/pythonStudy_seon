```python
piano = {'do': 261, 're':293, 'mi':329, 'pa':349, 'sol':391, 'ra':440, 'si':493, 'doo':530}
mel1 = ['do']
dur1 = [3]
music1 = zip(mel1,dur1)

mel2 = ['re']
dur2 = [3]
music2 = zip(mel2,dur2)

mel3 = ['mi']
dur3 = [3]
music3 = zip(mel3,dur3)

mel4 = ['pa']
dur4 = [3]
music4 = zip(mel4,dur4)

mel5 = ['sol']
dur5 = [3]
music5 = zip(mel5,dur5)

mel6 = ['ra']
dur6 = [3]
music6 = zip(mel6,dur6)

mel7 = ['si']
dur7 = [3]
music7 = zip(mel7, dur7)

mel8 = ['doo']
dur8 = [3]
music8 = zip(mel8, dur8)

for melody, duration in music1:
    duration = duration
while True:
    keyboard = input('')
    if (keyboard =='0'):
        break
    elif (keyboard == '1'):
        winsound.Beep(piano['do'], 1000//duration)
    elif (keyboard == '2'):
        winsound.Beep(piano['re'], 1000//duration)
    elif (keyboard == '3'):
        winsound.Beep(piano['mi'], 1000//duration)
    elif (keyboard == '4'):
        winsound.Beep(piano['pa'], 1000//duration)
    elif (keyboard == '6'):
        winsound.Beep(piano['sol'], 1000//duration)
    elif (keyboard == '7'):
        winsound.Beep(piano['ra'], 1000//duration)
    elif (keyboard == '8'):
        winsound.Beep(piano['si'], 1000//duration)
    elif (keyboard == '9'):
        winsound.Beep(piano['doo'], 1000//duration)
    continue
```

```python
import pygame

import pygame

pygame.init()

WHITE = (255, 255, 255)
font = pygame.font.SysFont("comicsansms", 72)
text = font.render("Byrobot Python Piano", True, (0, 128, 0))

pyscreen = pygame.display.set_mode((1024,768), 0, 32)
pyscreen.fill(WHITE)
pyscreen.blit(text,(512 - text.get_width() // 2, 384 - text.get_height() // 2))
pygame.display.flip()

run = True

c4 = pygame.mixer.Sound('music1')
d4 = pygame.mixer.Sound('music2')
e4 = pygame.mixer.Sound('music3')
f4 = pygame.mixer.Sound('music4')
g4 = pygame.mixer.Sound('music5')
a4 = pygame.mixer.Sound('music6')
b4 = pygame.mixer.Sound('music7')
c5 = pygame.mixer.Sound('music8')
d5 = pygame.mixer.Sound('music9')
# e5 = pygame.mixer.Sound('e5.wav')
# f5 = pygame.mixer.Sound('f5.wav')
# g5 = pygame.mixer.Sound('g5.wav')
# a5 = pygame.mixer.Sound('a5.wav')
# b5 = pygame.mixer.Sound('b5.wav')


while run:
    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                run = False

            elif event.key == pygame.K_u:
                c4.play()

            elif event.key == pygame.K_y:
                d4.play()

            elif event.key == pygame.K_t:
                e4.play()

            elif event.key == pygame.K_r:
                f4.play()

            elif event.key == pygame.K_e:
                g4.play()

            elif event.key == pygame.K_w:
                a4.play()

            elif event.key == pygame.K_q:
                b4.play()

            elif event.key == pygame.K_a:
                c5.play()

            elif event.key == pygame.K_s:
                d5.play()
                
            elif event.key == pygame.K_d:
                e5.play()

            elif event.key == pygame.K_f:
                f5.play()

            elif event.key == pygame.K_g:
                g5.play()

            elif event.key == pygame.K_h:
                a5.play()

            elif event.key == pygame.K_j:
                b5.play()
                
pygame.quit()
```
