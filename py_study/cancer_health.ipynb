# 간질환 파생변수생성

```python
def GPT_ALT(data):
    
    ga = data['감마지티피']/data['(혈청지오티)ALT']
    
    return ga

health_data['GPT_ALT'] = health_data.apply(GPT_ALT, axis=1)
```

```python
def AST_ALT(data):
    
    aa = data['(혈청지오티)AST']/data['(혈청지오티)ALT']
    
    return aa

health_data['AST_ALT'] = health_data.apply(AST_ALT, axis=1)
```

```python
def scoring_data(data):
    score = "0"
    
    if data['(혈청지오티)ALT'] <= 40:
        score = "00"

    elif data['(혈청지오티)ALT'] > 40 and data['(혈청지오티)ALT'] <= 200:
        if data['BMI'] > 20:
            score = "10"
        else:
            score = "11"

    elif data['(혈청지오티)ALT'] > 200 and data['(혈청지오티)ALT'] <= 400:
        if data['성별코드'] == 1:
            if data['감마지티피'] > 180:
                score = "21"
            else:
                score = "20"
        else:
            if data['감마지티피'] > 110:
                score = "21"
            else:
                score = "20"

    elif data['(혈청지오티)ALT'] > 400 and data['(혈청지오티)ALT'] <= 1000:
        score = "30"

    else:
        score = "31"
            
    if data['GPT_ALT'] > 2:
        if data['AST_ALT'] > 1.5:
            if data['(혈청지오티)AST'] > 160:
                score = "22"
            
    return score

health_data['간손상지수'] = health_data.apply(scoring_data, axis=1)
```

```python
#이웃 수 증가에 다른 정확도 변화

train_accuracy = []; test_accuracy = []
para_n_neighbors = [i for i in range(1,16)]

for v_n_neighbors in para_n_neighbors:
    knn = KNeighborsClassifier(n_neighbors = v_n_neighbors)
    knn.fit(df_train_x, df_train_y)
    train_accuracy.append(knn.score(df_train_x, df_train_y))
    test_accuracy.append(knn.score(df_test_x, df_test_y))
    
df_accuracy_neighbors = pd.DataFrame()
df_accuracy_neighbors['Neighbors'] = para_n_neighbors
df_accuracy_neighbors['TrainAccuracy'] = train_accuracy
df_accuracy_neighbors['TestAccuracy'] = test_accuracy
df_accuracy_neighbors.round(3)
```

```python
plt.plot(para_n_neighbors, train_accuracy, linestyle='-', label='Train Accuracy')
plt.plot(para_n_neighbors, test_accuracy, linestyle='--', label='Test Accuracy')
plt.ylabel('accuracy'); plt.xlabel('n_neighbors')
plt.legend()
```

```python
#가중치별 이웃 수 증가에 따른 정확도 변화

train_accuracy = []; test_accuracy = []
para_n_neighbors = [i for i in range(1,11)] * 2
para_weights = (['uniform']) * 10 + (['distance'] * 10)

for (v_n_neighbors, v_weights) in zip(para_n_neighbors, para_weights):
    knn = KNeighborsClassifier(n_neighbors = v_n_neighbors, weights = v_weights)
    knn.fit(df_train_x, df_train_y)
    train_accuracy.append(knn.score(df_train_x, df_train_y))
    test_accuracy.append(knn.score(df_test_x, df_test_y))
    
df_accuracy_weights = pd.DataFrame()
df_accuracy_weights['Neighbors'] = para_n_neighbors
df_accuracy_weights['Weights'] = para_weights
df_accuracy_weights['TrainAccuracy'] = train_accuracy
df_accuracy_weights['TestAccuracy'] = test_accuracy
df_accuracy_weights.round(3)
```

```python
def bar_chart(feature) :
    A = cehealth[cehealth['dang']==0][cehealth['sim']==0][cehealth['gan']==0][feature].value_counts()
    B = cehealth[cehealth['dang']==1][cehealth['sim']==0][cehealth['gan']==0][feature].value_counts()
    C = cehealth[cehealth['dang']==0][cehealth['sim']==1][cehealth['gan']==0][feature].value_counts()
    D = cehealth[cehealth['dang']==1][cehealth['sim']==1][cehealth['gan']==0][feature].value_counts()
    E = cehealth[cehealth['dang']==0][cehealth['sim']==2][cehealth['gan']==0][feature].value_counts()
    F = cehealth[cehealth['dang']==1][cehealth['sim']==2][cehealth['gan']==0][feature].value_counts()
    G = cehealth[cehealth['dang']==0][cehealth['sim']==0][cehealth['gan']==1][feature].value_counts()
    H = cehealth[cehealth['dang']==1][cehealth['sim']==0][cehealth['gan']==1][feature].value_counts()
    I = cehealth[cehealth['dang']==0][cehealth['sim']==1][cehealth['gan']==1][feature].value_counts()
    J = cehealth[cehealth['dang']==1][cehealth['sim']==1][cehealth['gan']==1][feature].value_counts()
    K = cehealth[cehealth['dang']==0][cehealth['sim']==2][cehealth['gan']==1][feature].value_counts()
    L = cehealth[cehealth['dang']==1][cehealth['sim']==2][cehealth['gan']==1][feature].value_counts()
    df = pd.DataFrame([A, B, C, D, E, F, G, H, I, J, K, L])
    df.index =['당뇨정상&혈압정상&간정상', '당뇨병&혈압정상&간정상', '당뇨정상&혈압주의&간정상' ,
               '당뇨병&혈압주의&간정상','당뇨정상&고혈압&간정상', '당뇨병&고혈압&간정상',
               '당뇨정상&혈압정상&간질환', '당뇨병&혈압정상&간질환', '당뇨정상&혈압주의&간질환',
               '당뇨병&혈압주의&간질환', '당뇨정상&고혈압&간질환', '당뇨병&고혈압&간질환']
    df.plot(kind='bar',stacked=True, figsize=(15,9))
```
